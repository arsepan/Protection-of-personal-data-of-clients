```python
# Создадим необходимые функции

def data_info(df):
    '''Информации о DF. Методы: describe, info, с просмотром памяти'''
    print(f'Размеры DF: {df.shape}', '\n')
    display(df.head(10), df.describe(), df.info(memory_usage='deep'))

def data_spread(column_list, df):
    '''Разброс данных. Отображение диаграммы размаха и гистограммы. Входные данные list колонок и df'''
    for column in column_list:
        df.boxplot(column=column)
        plt.title(f'Диаграмма размаха на столбце {column}')
        plt.show()
        df[column].plot(kind='hist', bins=100, grid=True)
        plt.title(f'Гистограмма на столбце {column}')
        plt.xlabel(column)
        plt.ylabel('Кол-во наблюдений')
        plt.show()
        print('--------------------------------------------')

def uniqueness(colomn_list, df):
    '''Просмотр уникальных значений листа'''
    for column in colomn_list:
        print(f'Количество уникальных значений столбца {column}: {df[column].nunique()}')
        print(df[column].unique())
        print('-----------------------------')
```

## Ответим на вопрос

Признаки умножают на обратимую матрицу. Изменится ли качество линейной регрессии?(Её можно обучить заново).

1. Изменится. Приведите примеры матриц.
2. Не изменится. Укажите, как связаны параметры линейной регрессии в исходной задаче и в преобразованной.

Что такое обратная матрица? Согласено определению, **обратная матрица - квадратная матрица с верхним индексом -1, которая при перемножении с обычной даёт единичную матрицу**. 

$$A*A^{-1}=A^{-1}*A=E$$

Нам необходимо понять **изменяется ли качество линейной регрессии, если признаки перемножить на обратную матрицу**. Вначале докажем формулой.

Вся задача линейной регрессии сводится к тому, чтобы найти вектор признаков и вектор весов и определить предсказание модели (также вектор). Всё это сильно напоминает задачи, которые мы решали в 5 классах $y = kx * b$ (регрессия у нас линейная, отсюда и схожесть с формулой. Мы ищем вектор, который будет максимально приближен ко всем точкам на координатной плоскости), только в нашем случае это выглядет немного иначе.

$$a = (x, w) + w_{0}$$, где


- a - целевая переменная
- x - матрица признаков(наблюдений), $n$ строк на $m + 1$ столбцов (включая фиктивную единичную колонку слева) с полным рангом по столбцам: $\text{rank}\left(X\right) = m + 1$
- w - вектор параметров модели, в машинном обучении эти параметры часто называют весами(параметрами)
- $w_{0}$ - сдвиг или, другими словами, случайная переменная, соответствующая случайной, непрогнозируемой ошибке модели

Для сокращения записи изменим обозначения. В матрицу X добавим столбец, состоящий только из единиц (он идёт нулевым); а параметр w0 — к вектору w, а затем умножим матрицу X на вектор w. Сдвиг умножится на вектор из единиц (нулевой столбец). Получим такой вектор предсказаний a

$$a = Xw$$

Один из способов вычислить значения параметров модели является **метод наименьших квадратов (МНК)**, который минимизирует среднеквадратичную ошибку между реальным значением зависимой переменной и прогнозом, выданным моделью:

$$w = (X^{T}*X)^{-1}*X^{T}*y$$

Вернёмся к нашему вопросу. Нам необходимо, чтобы работало равенство

$$a = a_{m}$$, где M - случайная обратная матрица

$$Xw = MXw_{m}$$

Пусть X1 у нас равен произведению матрицы признаков на рандомную матрицу 

$$X1 = X*M$$

Подставим X1 в $w = (X^{T}*X)^{-1}*X^{T}*y$, чтобы в дальнейшем решить уравнение $a = Xw$:

$$w = (X1^{T}*X1)^{-1}*X1^{T}*y$$

$$w = (X^{T}*M^{T}*X*M)^{-1}*X^{T}*M^{T}*y$$

$$w = (M^{T}*(X^{T}*X)*M)^{-1}*X^{T}*M^{T}*y$$

$$w = (M^{T})^{-1}*(X^{T}*X)^{-1}*(M)^{-1}*X^{T}*M^{T}*y$$

$$w = (M)^{-1}*w$$

Теперь подставим значение w в нашу формулу

$$a = X1M^{-1}w$$

$$a = XMM^{-1}w$$

$$a = Xw$$

**Ответ:** качество модели линейной регрессии не изменится. Мы доказали, что $a = a_{m}$, а параметры связаны $w = M^{-1}w$

## Алгоритм преобразования и его проверка

Итак, вернёмся к нашей главной задача: нужно защитить данные клиентов страховой компании «Хоть потоп» и разработать такой метод преобразования данных, чтобы по ним было сложно восстановить персональную информацию. По-моему мнению, отличным алгоритмом преобразования будет домножение признаков (данных клиентов) на обратную матрицу, такую, что $A*A^{-1}=E$, а также соответствующему условию $A_{m} = M_{n}$. Мы уже доказали, что качество модели будет неизменным с точки зрения математики, поэтому необходимо будет доказать это на качестве метрики R2, которая вплотную работает MSE. Пройдёмся мы по следующим этапам:

1. Разделим выборку обычной DF на таргет и трейн, а также разобьём их на тест и трейн
2. Обучить модель на стандартных данных клиента и записать метрику R2
3. Создадим матрицу(квадратная, размером равным столбцам нашим признаков), на которую в последствии мы будет домножать наши признаки
4. Перемножим наши признаки на созданную матрицу
5. Проверим новую R2 метрику на преобразованных данных
6. Сравним две R2 метрики

### Без преобразования

```python
# Разделим выборку на тест и траин, а также на таргет с признаками

target = data['Страховые выплаты']
features = data.drop('Страховые выплаты', axis=1)

fatures_train, features_test, target_train, target_test = train_test_split(features,
                                                                           target,
                                                                           test_size=.3,
                                                                           random_state=12345)

print(fatures_train.shape, features_test.shape, target_train.shape, target_test.shape)
```

```python
# Посмотрим как обучается линейная модель на данной выборке (маштабирование, кодирование и прочее не стал делать, 
# тк задача заключалась не в этом)

model = LinearRegression().fit(fatures_train, target_train)
normal_lr = r2_score(target_test, model.predict(features_test))
print(f'R2 = {normal_lr}')
```

### С преобразованием

```python
# Создадим матрицу с рандомными значениями и проверим её на обратимость

f = True

while f: 
    matrix = np.random.randint(1, 10, (features.shape[1], features.shape[1]))
    if np.linalg.det(matrix) != 0:
        f = False
        
print(matrix)    
```

```python
# Закодируем наши данные

encoded_matrix = features @ np.linalg.inv(matrix)
encode_features = pd.DataFrame(encoded_matrix.values, columns=features.columns)
display(data, encode_features)
```

```python
# Разделим выборку на тест и траин
features_train_enc, features_test_enc, target_train_enc, target_test_enc = train_test_split(encoded_matrix,
                                                                           target,
                                                                           test_size=.3,
                                                                           random_state=12345)

print(features_train_enc.shape, features_test_enc.shape, target_train_enc.shape, target_test_enc.shape)
```

```python
# Посмотрим на линейную регрессию на закодированных данных

model_enc = LinearRegression().fit(features_train_enc, target_train_enc)
encode_lr = r2_score(target_test_enc, model_enc.predict(features_test_enc))
print(f'R2 = {encode_lr}')
```

```python
# Сравним результат
ans = pd.DataFrame(data=[normal_lr, encode_lr], columns=['R2'], index=['Без преобразования', 'С преобразованием'])
ans
```
